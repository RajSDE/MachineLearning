```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Collecting the data from the Dataset

```python
diabetes_dataset = pd.read_csv('diabetes.csv') 
```

Printing the Dataset

```python
diabetes_dataset.head()
```

Shape of the data set - (rows,column)

```python
diabetes_dataset.shape
```

```python
diabetes_dataset['Outcome'].value_counts()
```

It means:

0--> Non Diabetic

1--> Diabetic

```python
diabetes_dataset.groupby('Outcome').mean()
```

Here we calculate the mean of all attributes refers to Diabetic and Non-Diabetic person

```python
# input
X = diabetes_dataset.drop(columns = 'Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

Data Standardization

```python
scaler = StandardScaler()
scaler.fit(X)
standardized_data = scaler.transform(X)
```

```python
X = standardized_data
```

```python
print(X)
print(Y)
```

Split the Data for Training and Testing the model

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

(614 - Dataset will be use for training)

(154 - Dataset will be use for testing)

Now, we will train our model

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, Y_train)
```

Model Evalution

For training

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy*100)
```

For testing

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```

We have train our Model to predict a person having Diabetes or Diabetes free
